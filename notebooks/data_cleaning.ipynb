```python
mask=df.duplicated()
duplicated=df[mask]

duplicated["Label"].value_counts(normalize=True)*100
```

Since FTP bruteforce and SSH bruteforce usually use automated scripting to conduct them , In my opinion they are not noise unlike benign ones which may be background noise etc

```python
#Dropping benign duplicated columns
mask= (duplicated["Label"]=="Benign")
to_drop=duplicated[mask]
duplicated.drop(to_drop.index,axis=0,inplace=True)
df.drop(to_drop.index,axis=0,inplace=True)
```

```python
duplicated["Label"].value_counts(normalize=True)*100
```

Successfuly dropped duplicated Benign noise 

We will need to normalize the dataset later on when building the neural network because it does affect weights alot 

Looking for features / columns that are mostly zeroes that kill model performance a lot


```python
nunique=df.nunique()
print(nunique.to_string())
const_cols=nunique[(nunique==1) | (nunique==2)]
print(len(const_cols))
```

```python
const_cols
```

The columns mentioned above have mostly a single value meaning they bring close to no variance to the model hence we drop these

Checking for negative infinity and positive infinity values

```python
print(np.isinf(df.select_dtypes(include="number")).sum().sum())
print(np.isinf(df.select_dtypes(include="number")).sum().to_string())

```

```python
num_cols = df.select_dtypes(include="number").columns
inf_cols = (np.isinf(df[num_cols])).sum()
print(inf_cols[inf_cols > 0])
```

```python
df[num_cols] = df[num_cols].replace([np.inf, -np.inf], 0)
```

Replaced +inf and -inf with 0 

```python
#sanity check
print(np.isinf(df.select_dtypes(include="number")).sum().sum())
print(np.isinf(df.select_dtypes(include="number")).sum().to_string())
```

Checking for Negative Values

```python
mask=((df["Init_Bwd_Win_Byts"]<0) | (df["Init_Bwd_Win_Byts"]<0))
new_df=df[mask]
new_df["Label"].value_counts(normalize=True)*100
```

```python
print(df["Label"].value_counts())
```

Most  of the -1 are Benign traffic meaning we can simply replace by 0 instead of dropping valuable columns


```python
df["Init_Fwd_Win_Byts"] = df["Init_Fwd_Win_Byts"].replace(-1, 0)
df["Init_Bwd_Win_Byts"] = df["Init_Bwd_Win_Byts"].replace(-1, 0)
```

```python
#Sanity check
num_col=df.select_dtypes(include="number")
print((num_col<0).sum())
```

```python
mask=((df["Fwd_IAT_Min"]<0) | (df["Fwd_IAT_Max"]<0) | (df["Flow_IAT_Min"]<0) | (df["Flow_IAT_Max"]<0) | (df["Fwd_IAT_Mean"]<0) | (df["Fwd_IAT_Tot"]<0) | (df["Flow_IAT_Mean"]<0) | (df["Flow_Pkts/s"]<0) )
new_df=df[mask]
new_df["Label"].value_counts(normalize=True)*100
```

All of these are benigns and prolly are errors mostly so we just drop em
